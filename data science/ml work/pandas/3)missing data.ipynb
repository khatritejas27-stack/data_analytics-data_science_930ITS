```python
import numpy as np
import pandas as pd
```

```python
#finding missing value
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df= pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
#removing null value
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=1)
```

```python
#filling missing data
df
```

```python
df.fillna(0)
```

```python
values={'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
