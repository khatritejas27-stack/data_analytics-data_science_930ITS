```python
import numpy as np
import pandas as pd
```

```python
#marging 2 dataframe
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
#marge dtaframe
pd.merge(employees,salaries,on='employee_id',how='inner')#not show nan value
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer') #show all nan value 
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')#marge bassis of name or
# department removing nan value
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')#marge bassis of salary or 
#bonus removing nan value
```

```python
#concantination of 2 dataframes
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```
